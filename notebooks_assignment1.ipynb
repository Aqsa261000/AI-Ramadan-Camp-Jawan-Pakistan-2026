```python
english = float(input("Enter your English marks: "))
mathematics = float(input("Enter your Mathematics marks: "))
computer=float(input("Enter your Computer marks: "))
urdu=float(input("Enter your Urdu marks: "))
islamiat=float(input("Enter your Islamiat marks: "))

total_marks=english+mathematics+computer+urdu+islamiat
percentage=(total_marks/500)*100

# grade for each subject
if(english>=30):
    grade1 = "Pass"
else:
    grade1 = "Fail"
if(mathematics>=30):
    grade2 = "Pass"
else:
    grade2 = "Fail"
if(computer>=30):
    grade3 = "Pass"
else:
    grade3 = "Fail"
if(urdu>=30):
    grade4 = "Pass"
else:
    grade4 = "Fail"
if(islamiat>=30):
    grade5 = "Pass"
else:
    grade5 = "Fail"

if(percentage>=80 and percentage<=100):
    grade = "A+"
elif(percentage>=70 and percentage<80):
    grade = "A"
elif(percentage>=60 and percentage<70):
    grade = "B"
elif(percentage>=50 and percentage<60):
    grade = "C"
elif(percentage>=40 and percentage<50):
    grade = "D"
elif(percentage>=30 and percentage<40):
    grade = "E"
else:
    grade = "F"


# Marksheet
print("\n--------------Marksheet--------------")
print("English marks:",english,"Grade:",grade1)
print("Mathematics marks:",mathematics,"Grade:",grade2)
print("Computer marks:",computer,"Grade:",grade3)
print("Urdu marks:",urdu,"Grade:",grade4)
print("Islamiat marks:",islamiat,"Grade:",grade5)
print("\nTotal marks:",total_marks)
print("Percentage:",percentage)
print("Grade:",grade)
```
