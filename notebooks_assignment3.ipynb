```python
firstNumber = float(input("Enter first number: "))
secondNumber = float(input("Enter second number: "))
operand=input("Enter the operation (+,-,*,/,**) you want to perform: ")

if(operand == '+'):
    result = firstNumber + secondNumber
    print("Your result is: ", result)
elif(operand == '-'):
    result = firstNumber - secondNumber
    print("Your result is: ", result)
elif(operand == '*'):
    result = firstNumber * secondNumber
    print("Your result is: ", result)
elif(operand == '**'):
    result = firstNumber ** secondNumber
    print("Your result is: ", result)
elif(operand == "/"):
    if(secondNumber!=0):
        result= firstNumber / secondNumber
        print("Your result is: ", result)
    else:
        print("Cannot divide by zero!")
else:
    print("You have entered an invalid operator")
```

```python
items = ["Aqsa",20,"Programmer"]
isNumeric = False
for item in items:
    if(type(item) == int or type(item) == float):
        print("Numeric value exists")
        isNumeric = True
        break;
if(isNumeric == False):
    print("No numeric value exists")
```

```python
fruits = ["Mango","Orange","Pomogrenate","Kiwi","Strawberry","Kiwi","Pomogrenate"]
seen=[]
duplicates=[]
for fruit in fruits:
    if fruit in seen:
        duplicates.append(fruit)
    else:
        seen.append(fruit)
        
print(duplicates,"are the duplicates in the fruits list")
```
